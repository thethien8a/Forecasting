```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    "Date": pd.date_range("2023-01-01", periods=13, freq="ME"),
    "Value": [10, 20, 30, 20, 30, 50, 40, 70, 90, 10, 14, 50, 80]
})
df["Date"] = pd.to_datetime(df["Date"])
df.set_index("Date", inplace=True)
```

```python
df["Value"].plot()
plt.show()
```

```python
ax = df.rolling(window=3).mean().dropna().plot(label='3-Month Rolling Mean', xlim=[df.index[0], df.index[-1]])
ax.set_ylim(0,100)
ax.set_title('3-Month Rolling Mean of Values')
ax.set_xlabel('Date')
ax.set_ylabel('Value')
plt.show()
```
